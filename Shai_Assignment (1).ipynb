### 3-Data cleaning

handel the missing values

first step use the dropna() method to drop all the columns have NAN value

then filling using the mean of each column .


```python
#salariesData_2 as a new salariesData after handling missing value
salariesData_2=salariesData.dropna(axis='columns', how='all')
salariesData_2
```

```python
salariesData_2.shape
```

as you see the number of columns have reduce to 11 (['Notes'] and ['Statues'] deleted)


```python
basepay_mean= salariesData_2['BasePay'].mean()
salariesData_2['BasePay'].fillna(value=basepay_mean, inplace=True)
```

```python
benefits_mean= salariesData_2['Benefits'].mean()
print(benefits_mean)
```

```python
salariesData_2['Benefits'].fillna(value=benefits_mean, inplace=True)
```

```python
overtimepay_mean= salariesData_2['OvertimePay'].mean()
salariesData_2['OvertimePay'].fillna(value= overtimepay_mean, inplace=True)
```

```python
otherpay_mean= salariesData_2['OtherPay'].mean()
salariesData_2['OtherPay'].fillna(value= otherpay_mean, inplace= True)

salariesData_2.head()
```

```python

salariesData_2.isnull().sum().sum()
```
